# ORL dataset results





This code  performs two tasks. Firstly, it downloads the ORL dataset from a Google Drive URL using the `gdown` library and saves it as "ORL.zip". Secondly, it demonstrates how to extract a RAR file using a function named `unrar_file`. This function is expected to extract the contents of a RAR file, 'ORL.rar' in this case, to a specified directory, 'ORL' here. Please note that the `unrar_file` function isn't defined in this snippet, so it should be defined elsewhere in your program.

```python
# Downlonlod ORL dataset

url = "https://drive.google.com/uc?id=1pYmYyU4R_56eZXe_uvmI8UORzbZvqBxL"
output = "ORL.rar"
gdown.download(url, output, quiet=False)

# Usage
path_to_rar_file = 'ORL.rar'
path_to_extract_to = 'ORL'
unrar_file(path_to_rar_file, path_to_extract_to)
```

**ORL dataset samples**

The ORL dataset contains 400 images of 40 subjects, each subject having 10 different images.
The images were taken at different times, varying the lighting, facial expressions (open / closed eyes, smiling / not smiling) and facial details (glasses / no glasses) The images are in grayscale and have a resolution of 92x112 pixels

```python

# Get the list of image files from the ORL dataset directory

imgorl = list(paths.list_images('ORL'))

# Display some random images from the ORL dataset
# The display_random_images function takes a list of image files as an argument and shows a specified number of images in a grid
# The function also prints the label of each image, which is extracted from the file name
display_random_images(imgorl)
```

**Feature extraction using optimised HOG and optimised image dimensions and train and evaluate**

```python
# Get the HOG features and labels from the ORL dataset

features, labels = extract_features_and_labels(imgorl)

# Train and evaluate the model using the features and labels
# The model uses the Ridge classifier with default hyperparameters
# The model is trained on 70% of the data and tested on 30% of the data
# The random_state argument is used to set the seed for the random number generator, ensuring reproducibility

train_and_evaluate_model(features, labels, random_state=0)
```
